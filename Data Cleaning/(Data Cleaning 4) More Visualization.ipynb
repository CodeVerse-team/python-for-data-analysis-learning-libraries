## More Visualizations

Previously, we saw an overview of how pandas `plot` method worked and how to use the basic API of matplotlib. We'll provide more details in this lesson.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## Global API

Matplotlib's default pyplot API has a global, MATLAB-style interface, as we've already seen:

```python
x = np.arange(-10, 11)
y = np.arange(-11, 12)
```

```python
plt.figure(figsize=(12, 6))

plt.title('My Nice Plot')

plt.plot(x, x ** 4)
plt.plot(y, y ** 2)
plt.plot(x, -1 * (x ** 4))
```

```python
plt.figure(figsize=(12, 6))
plt.title('Second Plot')

plt.subplot(1,2,1) # rows, columns, panel selected
plt.plot(x, x ** 2)
plt.plot([0,0,0], [-10, 0, 100])
plt.legend(['X^2', 'Vertical Line'])
plt.xlabel('X')
plt.ylabel('Y')

plt.subplot(1,2,2)
plt.plot(x, -1 * (x ** 2))
plt.plot([-10, 0, 10], [-50, -50, -50])
plt.legend(['-X^2', 'Horizontal line'])

plt.xlabel('X')
plt.ylabel('X Squared')
```

## OOP Interface

```python
fig, axes = plt.subplots(figsize = (12, 6))
```

```python
axes.plot(
    x, (x ** 2), color='red', linewidth=3,
    marker='o', markersize=5, label='X ^ 2')

axes.plot(x, -1 * (x ** 2), 'b--', label = '-X^2')

axes.set_title('Third Plot Using OOP')
axes.legend()
fig
```

```python
fig, axes = plt.subplots(figsize = (12, 6))

axes.plot(x, x + 0, linestyle = 'solid')
axes.plot(x, x + 1, linestyle = 'dashed')
axes.plot(x, x + 2, linestyle = 'dashdot')
axes.plot(x, x + 3, linestyle = 'dotted');

axes.set_title("Fourth Plot")
```

```python
fig, axes = plt.subplots(figsize = (12, 6))

axes.plot(x, x + 0, '-og', label = "solid green")
axes.plot(x, x + 1, '--c', label = "dashed cyan")
axes.plot(x, x + 2, '--y', label = "dashdot yellow")
axes.plot(x, x + 3, ':r', label = "dotted red")

axes.set_title("Fifth Plot")
axes.legend()
```

There are lot of line and marker types.

```python
print('Markers: {}'.format([m for m in plt.Line2D.markers]))
```

```python
lineStyles = ['_', '-', '--', ':']
print('Line styles: {}'.format(lineStyles))
```

## Other types of plots

### Figures and subfigures

When we call the `subplots()` function we get a tuple containing a `Figure` and `axes` element.

```python
plot_objects = plt.subplots()

fig, ax = plot_objects
ax.plot([1,2,3], [1,2,7]) # [1,2,3] [1,2,3] real values
plot_objects
```

We can also define how many elements we want inside our figure. To do that we can set the `nrows` and `ncols` params

```python
plot_objects = plt.subplots(nrows=2, ncols=2, figsize=(14, 6))

fig, ((ax1, ax2), (ax3, ax4)) = plot_objects

plot_objects
```

```python
ax1.plot(np.random.randn(50), c = 'red', linestyle = '--')
ax2.plot(np.random.randn(50), c = 'green', linestyle = ':')
ax3.plot(np.random.randn(50), c = 'blue', marker = 'o', linewidth = 1.0)
ax4.plot(np.random.randn(50), c = 'yellow')


fig
```

## The `Subplot2grid` command
There is another way to make subjects using a grid like format:

```python
plt.figure(figsize=(14, 6))

ax1 = plt.subplot2grid((3, 3), (0, 0), colspan = 3)
ax2 = plt.subplot2grid((3, 3), (1, 2), colspan = 2)
ax3 = plt.subplot2grid((3, 3), (1, 2), rowspan = 2)
ax4 = plt.subplot2grid((3, 3), (2, 0))
ax5 = plt.subplot2grid((3, 3), (2, 1))
```

## Scatter Plot

```python
N = 60
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (20 * np.random.rand(N)) ** 2 # 0 to 15 point radii
```

```python
plt.figure(figsize = (14, 6))

plt.scatter(x, y, s = area, c = colors, alpha = 0.5, cmap = 'Spectral')
plt.colorbar()

plt.show()
```

```python
fig = plt.figure(figsize=(14, 6))

ax1  = fig.add_subplot(1, 2, 1)
plt.scatter(x, y, s = area, c = colors, alpha = 0.6, cmap = 'spring')
plt.colorbar()

ax2  = fig.add_subplot(1, 2, 2)
plt.scatter(x, y, s = area, c = colors, alpha = 0.8, cmap = 'GnBu')
plt.colorbar()

plt.show()
```

Here is the full `cmap` options available: [https://matplotlib.org/2.0.2/users/colormaps.html](https://matplotlib.org/2.0.2/users/colormaps.html)

## Histograms

```python
values = np.random.randn(1000)
```

```python
plt.subplots(figsize = (12, 6))

plt.hist(values, bins = 100, alpha = 0.5,
         histtype = 'bar', color = 'steelblue',
         edgecolor = 'green')
plt.xlim(xmin = -5, xmax = 5)

plt.show()
```

```python
fig.savefig('hist.png')
```

## KDE (kernel density estimation)

```python
from scipy import stats

density = stats.kde.gaussian_kde(values)
density
```

```python
plt.subplots(figsize = (12, 6))

values2 = np.linspace(min(values)-10, max(values)+10, 100)

plt.plot(values2, density(values2), color='#FF7F00')
plt.fill_between(values2, 0, density(values2), alpha = 0.5, color='#FF7F00')
plt.xlim(xmin = -5, xmax = 5)

plt.show()
```

## Combine Plots

```python
plt.subplots(figsize = (12, 6))

plt.hist(values, bins = 100, alpha = 0.8, density = 1,
         histtype= 'bar', color = 'green',
         edgecolor = 'blue')

plt.plot(values2, density(values2), color='#FF7F00')
plt.fill_between(values2, 0, density(values2), alpha = 0.5, color='#FF7F00', linewidth=3.0)
plt.xlim(xmin = -5, xmax = 5)

plt.show()
```

## Bar Plots

```python
Y = np.random.rand(1, 5)[0]
Y2 = np.random.rand(1, 5)[0]
```

```python
plt.figure(figsize = (12, 4))

barWidth = 0.5
plt.bar(np.arange(len(Y)), Y, width = barWidth, color = '#00b894')

plt.show()
```

Also can be stacked bars, and add a legend to the plot:

```python
plt.figure(figsize = (12, 4))

barWidth = 0.5
plt.bar(np.arange(len(Y)), Y, width=barWidth, color = '#00b894', label = 'label Y')
plt.bar(np.arange(len(Y2)), Y, width=barWidth, color = '#e17055', bottom = Y, label = 'label Y2')

plt.legend()
plt.show()
```
